```python
import random
import matplotlib.pyplot as plt

def GISimulation():
    tunnel_len = 50
    imp_position = 5
    golem_position = 0

    imp_positions = [imp_position]
    golem_positions = [golem_position]

    print("Pursuit begins!")
    seconds = 0
    while imp_position < tunnel_len and golem_position < tunnel_len:
        imp_move = random.randint(1, 9)
        golem_move = random.randint(3, 5)

        imp_position += imp_move
        golem_position += golem_move

        imp_positions.append(imp_position)
        golem_positions.append(golem_position)

        
        
        seconds += 1
        if imp_position >= tunnel_len:
            print("Imp has escaped to freedom after " + str(seconds) + " seconds. Mischief and trickery await!")
            break
        elif golem_position >= imp_position:
            print("Sadly, it's back to the tower for the imp after a " + str(seconds) + " second chase.")
            break

        print("-" * golem_position + "G" + "-" * (imp_position - golem_position - 1) + "I" + "-" * (tunnel_len - imp_position - 1) + "X\t" + f"{seconds}s")
    
    # Visualize the chase
    plt.plot(range(len(imp_positions)), imp_positions, label="Imp")
    plt.plot(range(len(golem_positions)), golem_positions, label="Golem")
    plt.xlabel("Time (seconds)")
    plt.ylabel("Position (feet)")
    plt.legend()
    plt.title("Imp-Golem Chase Simulation")
    plt.show()

# Run the simulation
GISimulation()
```
