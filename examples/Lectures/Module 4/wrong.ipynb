```python
def cube(x):
    """Calculate the cube of x"""
    r = x ** 3
    return r
print('The cube of 2 is', cube(2))
```

```python
def seconds_past_midnight(hour, minute, second):
    """Calculate the number of seconds past midnight"""
    return hour * 3600 + minute * 60 + second
print('The number of seconds past midnight is', seconds_past_midnight(4, 5, 25))
```

```python
import matplotlib.pyplot as plt
def celtofah(celtemp):
    """Convert Celsius to Fahrenheit"""
    fah = celtemp * 9 / 5 + 32
    return fah

cellist = []
fahlist = []
for celsius in range(0, 101, 5):
    cellist.append(celsius)
    fahlist.append(celtofah(celsius))

plt.bar(cellist, fahlist, align = 'center', width=2)
plt.xlabel('Degrees (Celcius)')
plt.ylabel('Degrees (Fahrenheit)')
```
