```python
import random
import matplotlib.pyplot as plt

def imp_golem_chase_simulation():
    tunnel_length = 50
    imp_position = 5
    golem_position = 0

    imp_positions = [imp_position]
    golem_positions = [golem_position]

    print("Chase begins!")

    while imp_position < tunnel_length and golem_position < tunnel_length:
        imp_move = random.randint(1, 9)
        golem_move = random.randint(3, 5)

        imp_position += imp_move
        golem_position += golem_move

        imp_positions.append(imp_position)
        golem_positions.append(golem_position)

        time_in_seconds = imp_position // 3
        print("-" * golem_position + "G" + "-" * (imp_position - golem_position - 1) + "I" + "-" * (tunnel_length - imp_position - 1) + "X\t" + f"{time_in_seconds}s")

        if imp_position >= tunnel_length:
            print("Imp has escaped! Freedom at last!")
        elif golem_position >= imp_position:
            print("Golem caught the imp! Back to the tower for the imp.")
            break
    
    # Visualize the chase
    plt.plot(range(len(imp_positions)), imp_positions, label="Imp")
    plt.plot(range(len(golem_positions)), golem_positions, label="Golem")
    plt.xlabel("Time (seconds)")
    plt.ylabel("Position (feet)")
    plt.legend()
    plt.title("Imp-Golem Chase Simulation")
    plt.show()

# Run the simulation
imp_golem_chase_simulation()
```
